```python
def timedomain(hr):   
    results = {}      
   
    results['Mean HR (beats/min)'] = np.mean(hr)
    results['STD HR (beats/min)'] = np.std(hr)
    results['Min HR (beats/min)'] = np.min(hr)
    results['Max HR (beats/min)'] = np.max(hr)
    
    return results

timedomain(Heart_Rate[147:220])
```

```python
def timedomain(rr):   #RR interval of a heartbeat in milliseconds
    results = {}      
   
    results['Mean RR (ms)'] = np.mean(rr)
    results['STD RR/SDNN (ms)'] = np.std(rr)  #Standard Deviation of NN intervals
    results['RMSSD (ms)'] = np.sqrt(np.mean(np.square(np.diff(rr))))   #root mean square of successive differences between RR intervals
    results['NNxx'] = np.sum(np.abs(np.diff(rr)) > 50)*1   #calculates the number of pairs of successive RR intervals that differ by more than 50 milliseconds
    results['pNNxx (%)'] = 100 * np.sum((np.abs(np.diff(rr)) > 50)*1) / len(rr)   #calculates the percentage of pairs of successive RR intervals that differ by more than 50 milliseconds relative to the total number of RR intervals
    return results

timedomain(rr_interval[147:220])
```
